### W5A16 Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def amstrong(x):
    sum = 0
    digits = len(str(x))
    temp = x
    while temp > 0:
        digit = temp % 10
        sum += digit**digits 
        temp //= 10
    if sum == x:
        return True
    return False
x = int(input())
print(amstrong(x))
```

### W5A17 Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def so_chinh_phuong(n):
    if n<0 :
        return False
    else:
        for i in range(1,n+1):
            if i*i == n:
                return True
        return False
n = int(input())
print(so_chinh_phuong(n))
```

### W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
def goc_abc(x1,y1,x2,y2,x3,y3):
    tu = (x1-x2)*(x3-x2) + (y1-y2)*(y3-y2)
    ba = math.sqrt((x1-x2)**2 + (y1-y2)**2)
    bc = math.sqrt((x3-x2)**2 + (y3-y2)**2)
    mau = ba*bc
    return math.degrees(math.acos(tu/mau))
x1,y1,x2,y2,x3,y3 = map(int, input().split())
print(goc_abc(x1,y1,x2,y2,x3,y3))
```

### W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0
    tong = 0
    #else:
    for i in range(1,n+1):
        s = str(i)
        if len(s) == len(set(s)):
            tong += i
    return tong
n = int(input())
print(sum_1_to_n(n))
```
